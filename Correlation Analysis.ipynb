```python
path = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'
df = pd.read_csv(path)
df.head(10)
```

```python
df_group1 = df[['drive-wheels','num-of-doors','price']]
df_group1 = df_group1.groupby(["drive-wheels"], as_index=False).mean()
df_group1
```

```python
df_group2 = df[['drive-wheels','num-of-doors','price']]
df_group2 = df_group2.groupby(['drive-wheels','num-of-doors'], as_index=False).mean()
df_group2
```

```python
df_group3 = df[['drive-wheels','body-style','price']]
```

```python
df_group3 = df_group3.groupby(['drive-wheels','body-style'], as_index=False).mean()
df_group3
```

```python
pivot1 = df_group3.pivot(index= "drive-wheels", columns = "body-style")
pivot1
```

```python
import numpy as np
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(pivot1, cmap='RdBu')

#label names
row_labels = pivot1.columns.levels[1]
col_labels = pivot1.index

#move ticks and labels to the center
ax.set_xticks(np.arange(pivot1.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(pivot1.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

```python
#Pearson Correlation

from scipy import stats
```

```python
pearson_coef , p_value = stats.pearsonr(df['horsepower'], df['price'])
print ('Pearson Correlation is', pearson_coef, 'P_Value is', p_value)
```

```python
spearman_coef , p_value = stats.spearmanr(df['horsepower'], df['price'])
print ('Spearman Correlation is', spearman_coef, 'P_Value is', p_value)
```
